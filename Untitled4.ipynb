```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
```

```python
df = pd.read_csv("HR_comma_sep.csv")
```

```python
print(df.head())
print(df.info())
print(df.describe())
print("Missing values per column:")
print(df.isnull().sum())
```

```python
sns.countplot(x="left", data=df)
plt.title("Count of employees who left vs stayed")
plt.show()
```

```python
ct = pd.crosstab(df["salary"], df["left"], normalize="index")
ct.plot(kind="bar", stacked=True)
plt.ylabel("Proportion")
plt.title("Salary level vs probability of leaving (proportions)")
plt.legend(title="Left (1 = left, 0 = stayed)")
```

```python
ct2 = pd.crosstab(df["Department"], df["left"], normalize="index")
ct2.plot(kind="bar", stacked=True, figsize=(10,5))
plt.ylabel("Proportion")
plt.title("Department vs probability of leaving")
plt.legend(title="Left (1 = left, 0 = stayed)")
plt.show()
```

```python
corr = df.drop(['Department', 'salary'], axis=1).corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation matrix")
plt.show()
```

```python
df2 = pd.get_dummies(df, columns=["salary", "Department"], drop_first=True)
```

```python
features = [
    "satisfaction_level",
    "last_evaluation",
    "number_project",
    "average_montly_hours",
    "time_spend_company",
    "Work_accident",
    "promotion_last_5years",
    # salary and department dummies:
] + [col for col in df2.columns if col.startswith("salary_")] \
  + [col for col in df2.columns if col.startswith("sales_")]
```

```python
X = df2[features]
y = df2["left"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:,1]
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification report:\n", classification_report(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_proba))
```

```python
coef_df = pd.DataFrame({
    "feature": X_train.columns,
    "coef": model.coef_[0]
})
coef_df = coef_df.sort_values(by="coef", ascending=False)
print(coef_df.head(10))
print(coef_df.tail(10))
```
